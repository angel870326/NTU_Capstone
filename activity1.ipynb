## 活動新大類

```python
# 節慶活動
festival = ['10007','10008','10009','10010','10011','10012','20005','20007','20009','20025','20027','20029','A0004','A0012','C0002','C0006','C0012','H0003','H0004','H0005','H0006','H0007','X0010']
# 聯歡
party = ['C0003','C0004','C0005','D0003']
# 園遊會
fair = ['D0001','D0002']
```

## Read file

```python
data = pd.read_excel('/content/gdrive/Shareddrives/110-2 Capstone/社區活動/資料檔/整理完的/合併後活動檔.xlsx')

# Number of observations and variables
print(data.shape)
```

## Functions

```python
# Concatenate cells(features) into a string with separator (ignore NA)
def concatenate(data_subset):
  df = pd.DataFrame(data_subset.c0101)
  df['content'] = data_subset.loc[:,'f0114':'f0158'].apply(lambda x: ','.join(x.dropna().values.tolist()), axis=1)
  labels = np.array(df['c0101'])
  return df

# Extract all the contents from a concatenated data_subset
def textList(c_data_subset):
  text_list = c_data_subset.iloc[0,1]
  for i in range(1,len(c_data_subset.index)):
    text_list = text_list + ',' + c_data_subset.iloc[i,1]
  text_list = text_list.split(',')
  return text_list
```

## 節慶活動

```python
festival_filter = data['c0103'].isin(festival)
data_festival = data.loc[festival_filter,:]
print(data_festival.shape) # Number of observations and variables
```

```python
dict_festival = Counter(textList(concatenate(data_festival)))
festival_total = sum(dict_festival.values())
print("Festival feature sum: ", festival_total)
festival_feature = pd.DataFrame(list(dict_festival.items()), columns=['Feature', 'Count']).sort_values(by=['Count'], ascending=False)
festival_feature['Weight'] = festival_feature['Count']/festival_total
festival_feature
```

## 聯歡

```python
party_filter = data['c0103'].isin(party)
data_party = data.loc[party_filter,:]
print(data_party.shape) # Number of observations and variables
```

```python
dict_party = Counter(textList(concatenate(data_party)))
party_total = sum(dict_party.values())
print("Party feature sum: ", party_total)
party_feature = pd.DataFrame(list(dict_party.items()), columns=['Feature', 'Count']).sort_values(by=['Count'], ascending=False)
party_feature['Weight'] = party_feature['Count']/party_total
party_feature
```

## 園遊會

```python
fair_filter = data['c0103'].isin(fair)
data_fair = data.loc[fair_filter,:]
print(data_fair.shape) # Number of observations and variables
```

```python
dict_fair = Counter(textList(concatenate(data_fair)))
fair_total = sum(dict_fair.values())
print("Fair feature sum: ", fair_total)
fair_feature = pd.DataFrame(list(dict_fair.items()), columns=['Feature', 'Count']).sort_values(by=['Count'], ascending=False)
fair_feature['Weight'] = fair_feature['Count']/fair_total
fair_feature
```

## 加上活動新大類欄位

```python
data_festival['type'] = '節慶活動'
data_party['type'] = '聯歡'
data_fair['type'] = '園遊會'
output = pd.concat([data_festival,data_party,data_fair])
print(output.shape) # Number of observations and variables
output
```
