## **New Taipei**

```python
activity_newtaipei = pd.read_excel('/content/gdrive/Shareddrives/110-2 Capstone/社區活動/社區活動特徵結果/activity_concat_newtaipei.xlsx')
cluster_newtaipei = pd.read_csv('/content/gdrive/MyDrive/碩一下/Capstone/hierarchical_clustering/taipei/hierarchical_textcluster_newtaipei.csv')

# Number of observations and variables
print("活動資料檔 (activity)：", activity_newtaipei.shape)
print("分群結果檔 (cluster)： ", cluster_newtaipei.shape)
```

### Functions

```python
# Array of building IDs for a cluster
def clusterFilterArray(cluster_city, method, clusterID):
  filter = cluster_city[method].isin([clusterID])
  method_clusterID = cluster_city.loc[filter,['c0101',method]]
  print(method, "-", clusterID, ":", method_clusterID.shape) # Number of observations and variables for a cluster
  return method_clusterID.c0101.to_numpy()

# Activities in a cluster
def activityInCluster(activity_city, cluster_buildingIDs):
  filter = activity_city['c0116'].isin(cluster_buildingIDs)
  nt_activity_method_cluster = activity_city.loc[filter,:]
  return nt_activity_method_cluster
```

### Ward 3

```python
# Create an array of building IDs for each cluster
print("各群建物數 (number of buildings in a cluster)：")
nt_ward_3_1_filter = clusterFilterArray(cluster_newtaipei,'cluster_ward_3',1)
nt_ward_3_2_filter = clusterFilterArray(cluster_newtaipei,'cluster_ward_3',2)
nt_ward_3_3_filter = clusterFilterArray(cluster_newtaipei,'cluster_ward_3',3)

print("\nExample array:", nt_ward_3_1_filter)
```

```python
# Find the activities held for each cluster
print("各群過去幾年辦的活動數：")
# Cluster 1
nt_activity_ward_3_1 = activityInCluster(activity_newtaipei, nt_ward_3_1_filter)
print(nt_activity_ward_3_1.shape) # Number of observations and variables
# Cluster 2
nt_activity_ward_3_2 = activityInCluster(activity_newtaipei, nt_ward_3_2_filter)
print(nt_activity_ward_3_2.shape) # Number of observations and variables
# Cluster 3
nt_activity_ward_3_3 = activityInCluster(activity_newtaipei, nt_ward_3_3_filter)
print(nt_activity_ward_3_3.shape) # Number of observations and variables
```

### Add cluster ID column

```python
nt_activity_ward_3_1['text_ward_3'] = '1'
nt_activity_ward_3_2['text_ward_3'] = '2'
nt_activity_ward_3_3['text_ward_3'] = '3'
nt_activity = pd.concat([nt_activity_ward_3_1, nt_activity_ward_3_2, nt_activity_ward_3_3])
print(nt_activity.shape) # Number of observations and variables
nt_activity
```
